```python
import seaborn as sns
import matplotlib

position_labels = player_dataset["Best Position"].unique()
rgb_values = sns.color_palette("Set2", len(position_labels))
position_color_map = dict(zip(position_labels, rgb_values))

plt.scatter(player_dataset["Passing Total"], player_dataset["Vision"], c=player_dataset["Best Position"].map(position_color_map), alpha=0.7)
plt.legend(handles=[matplotlib.patches.Patch(color=position_color_map[label], label=label) for label in position_labels])
```
